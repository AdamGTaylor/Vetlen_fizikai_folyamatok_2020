```python
import matplotlib.pyplot as plt
import numpy as np
import random

import sys 
import os
```

### Paraméterek

```python
#defining static stuff
k_B = 1.380648e-23
J = 5e-22
params = [0.09, 0.18, 0.50, 0.95]
T1 = J / k_B / params[0]
T2 = J / k_B / params[1]
T3 = J / k_B / params[2]
T4 = J / k_B / params[3]

Ts = [T1,T2,T3,T4]
betas = [1/k_B/T1,1/k_B/T2,1/k_B/T3,1/k_B/T4]

print(T1, " K\n", T2, " K\n", T3, " K\n", T4, " K\n")
print(betas[0], "\n", betas[1], "\n", betas[2], "\n", betas[3], "\n")
```

### Függvény, amivel egy random állapotba állnak be majd a spinek

```python
def chooose_sign():
    return -1 if random.random() < 0.5 else 1
```

### A teljes energia meghatározás egy adott pillanatban (állapotban)
Ezt a feladat meghatározásának megfelelően

```python
def E_1D(Spins):
    E = 0
    for i in range(0, len(Spins)-1):
        E += Spins[i] * Spins[i+1]
    return -E * J
```

### Léptetése a szimulációnak
1. választunk egy spint
2. $\Delta$E számítása
3. HA $\Delta$E < 0, akkor forgatjuk a spint, és vissza (1)-re
4. HA $\Delta$E = 0, akkor egy random szám húzása [0;1] intervallumból, ami ha kisebb mint 1/2 akkor forgatjuk, és vissza (1)-re. HA nagyobb 1/2-nél, nincs forgatás, vissza (1)-re. ((MI VAN HA PONT 1/2???))
5. HA $\Delta$E > 0, akkor véletlen szám [0,1]-ból, és ha az kisebb, mint exp(-$\beta \Delta$ E), akkor forgatjuk, egyébként megyünk (1)-re

```python
def simulation_1D(Beta, Steps, Spins):
    
    spin_changes = np.zeros((Steps+1,len(Spins)), dtype=int)
    energy = np.zeros((Steps+1))
    total_magnetization = np.zeros((Steps+1))
    
    #let's start it
    spin_changes[0] = Spins               #alapállapotban nincs változás
    energy[0] = E_1D(Spins = Spins)          #spinekkel alapállapoti energia
    total_magnetization[0] = Spins.sum()/len(Spins)
    
    #haladjunk!
    for current_step in range(0,Steps):
        choose_random_spin = np.random.choice([i for i in range(0,len(Spins))])
        
        #először alkossuk meg a frgatott állapotot
        new_Spins = Spins
        new_Spins[choose_random_spin] *= -1 #ez egy feltétele új állapot
        #a feltételes új állapot energiája
        E_1 = E_1D(Spins=Spins)
        E_2 = E_1D(Spins=new_Spins)
        delta_E = E_2 - E_1
        #most nézzük meg, hogy forduljon e fenti leírás alapján
        #3.
        if(delta_E < 0):
            spin_changes[current_step+1] = new_Spins
            Spins = New_Spins
            energy[current_step+1] = E_2
        #4.
        if (delta_E == 0):
            if(random.random() < 0.5):
                #vált
                spin_changes[current_step+1] = new_Spins
                Spins = new_Spins
                energy[current_step+1] = E_2
            else:
                #marad
                spin_changes[current_step+1] = Spins
                energy[current_step+1] = E_1
        else: #tehát nagyobb mint nulla deltaE
            if(random.random() < np.exp(-beta * delta_E)):
                spin_changes[current_step+1] = new_Spins
                Spins = new_Spins
                energy[current_step + 1] = E_2
            else:
                spins_changes[current_step+1] = Spins
                energy[current_step+1] = E_1
        #valamelyik kimenet megvan, kkor számolható a mágnesezettség        
        total_magnetization[current_step] = Spins.sum()/len(Spins)
    
    #kiment
    return spin_changes, energy, total_magnetization, Spins
```

Most hogy megvan a szimuláció, ami végrehajtja a lépéseket, ezért érdemes mint a négy `beta`n végigmenni, valamilyen lépésszám és spin darabszám mellett)ez a feladat is).
```python
#oké, ha megvan az alapállapot, akkor mehet a szimuláció, persze kellenek
#adattárolók is, mivel nem szeretném össszezavarni sem magam, se senki mást,
#ezért külön külön adatstruktúrát csinálok
n_Spins = 50
SPINS = np.array([chooose_sign() for i in range(0,n_Spins)])
STEPS = 2000

spin_changes_list = np.zeros((len(betas), STEPS+1,len(SPINS)))
energies = np.zeros((len(betas), STEPS+1))
tot_mags = np.zeros((len(betas), STEPS+1))
SPINS_S = np.zeros((len(betas), len(SPINS)))
print(spin_changes_list.shape, energies.shape, tot_mags.shape, SPINS_S.shape)


for i in range(0,len(betas)):
    spin_changes_list[i] , energies[i], tot_mags[i], SPINS_S[i] = simulation_1D(betas[i], Steps=STEPS, Spins = SPINS.copy())
    
indeces = np.arange(0,STEPS+1)
```

```python
#akkor most mehetnek az első plotok is
#nézzük  az energia változása
SIZE = int(len(betas)/2)
lapult_x = 400

fig, ax  = plt.subplots(nrows=SIZE, ncols=SIZE, figsize=(20,20))

for i in range(0,SIZE):
    for j in range(0,SIZE):
        #ax[i][j].axvline(x=lapult_x, color='black', linestyle="--", label="Az egyensúlyi állapot kezdete")
        ax[i][j].axhline(y=0, color='blue', linestyle="-", label="Egyensúly")
        ax[i][j].plot(indeces, energies[j+i*SIZE],color="cyan", linestyle="-", label="Energia nagysága egy állapotban")
        ax[i][j].set_title("Energia változása \n T=" + str(Ts[j+i*SIZE]) + " K", fontsize=30)
        ax[i][j].set_xlabel("Idő [N$\\cdot \\tau$]", fontsize=25)
        ax[i][j].set_ylabel("Energia [J]", fontsize=30)
        ax[i][j].tick_params(axis='both', which='major', labelsize=20)
        ax[i][j].legend(loc='upper right',fontsize=20)
        ax[i][j].yaxis.get_offset_text().set_fontsize(15)
        ax[i][j].grid()
fig.tight_layout()
plt.savefig('img\\energy_time.pdf', format='pdf', dpi=150, bbox="tight")
plt.show()
```

Látható a hasonló viselkedés ebben is. De ha ez az egyensúlyhoz tart, akkor megjósolható, hogy a várható értéke is 0-hoz tart. De egy ábra beszédesebb!

```python
avg_tot_mags2 = np.zeros((len(betas), STEPS+1))
for j in range(0,len(betas)):
    for i in range(1,STEPS+2):
        avg_tot_mags2[j,i-1] = sum(tot_mags2[j,:i])/i
        
fig, ax  = plt.subplots(nrows=SIZE, ncols=SIZE, figsize=(20,20))

for i in range(0,SIZE):
    for j in range(0,SIZE):
        ax[i][j].plot(indeces, avg_tot_mags2[j+i*SIZE],color="magenta", linestyle="-", label="<m>")
        ax[i][j].set_title("<m>\n T=" + str(Ts[j+i*SIZE]) + " K", fontsize=30)
        ax[i][j].set_xlabel("Idő [N$\\cdot \\tau$]", fontsize=25)
        ax[i][j].set_ylabel("<m>", fontsize=30)
        ax[i][j].tick_params(axis='both', which='major', labelsize=20)
        ax[i][j].legend(loc='lower right',fontsize=20)
        ax[i][j].yaxis.get_offset_text().set_fontsize(15)
        ax[i][j].grid()
fig.tight_layout()
plt.savefig('img\\m_avg_time.pdf', format='pdf', dpi=150, bbox="tight")
plt.show()
```

Az ábrákon keresztül is beláttuk ezt. Most nézük meg a $<m^2>$ mennyiséget is!

```python
avg_tot_mags2_sq = np.zeros((len(betas), STEPS+1))
for j in range(0,len(betas)):
    for i in range(1,STEPS+2):
        avg_tot_mags2_sq[j,i-1] = sum((tot_mags2[j,:i])**2)/i
        
fig, ax  = plt.subplots(nrows=SIZE, ncols=SIZE, figsize=(20,20))

for i in range(0,SIZE):
    for j in range(0,SIZE):
        ax[i][j].plot(indeces, avg_tot_mags2_sq[j+i*SIZE],color="magenta", linestyle="-", label="<m$^2$>")
        ax[i][j].set_title("<m$^2$>\n T=" + str(Ts[j+i*SIZE]) + " K", fontsize=30)
        ax[i][j].set_xlabel("Idő [N$\\cdot \\tau$]", fontsize=25)
        ax[i][j].set_ylabel("<m$^2$>", fontsize=30)
        ax[i][j].tick_params(axis='both', which='major', labelsize=20)
        ax[i][j].legend(loc='upper right',fontsize=20)
        ax[i][j].yaxis.get_offset_text().set_fontsize(15)
        ax[i][j].grid()
fig.tight_layout()
plt.savefig('img\\m_sq_avg.pdf', format='pdf', dpi=150, bbox="tight")
plt.show()
```

Valamennyire várható volt, hogy a rendszer ilyen viselkedéső lesz, miszen ha m várható értéke nullához tart, akkor $m^2$ várható értéke is oda fog tartani és nem mutatok meg, de m várható értékének négyzete is.

Már csak egy dolgot kell megnéznünk, a $<m^2> - <m>^2 = \sigma^2$ mennyiséget

## Szórás négyzete, avagy : $<m^2> - <m>^2 = \sigma^2$

```python
fig, ax  = plt.subplots(nrows=SIZE, ncols=SIZE, figsize=(20,20))

for i in range(0,SIZE):
    for j in range(0,SIZE):
        ax[i][j].plot(indeces, (avg_tot_mags2_sq[j+i*SIZE] - avg_tot_mags2[j+i*SIZE]**2),color="magenta", linestyle="-", label="<m$^2$> - <m>$^2$")
        ax[i][j].set_title("<m$^2$> - <m>$^2$\n T=" + str(Ts[j+i*SIZE]) + " K", fontsize=30)
        ax[i][j].set_xlabel("Idő [N$\\cdot \\tau$]", fontsize=25)
        ax[i][j].set_ylabel("<m$^2$> - <m>$^2$", fontsize=30)
        ax[i][j].tick_params(axis='both', which='major', labelsize=20)
        ax[i][j].legend(loc='lower right',fontsize=20)
        ax[i][j].yaxis.get_offset_text().set_fontsize(15)
        ax[i][j].grid()
fig.tight_layout()
plt.savefig('img\\var_time.pdf', format='pdf', dpi=150, bbox="tight")
plt.show()
```
